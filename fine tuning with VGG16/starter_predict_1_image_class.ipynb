```python
#libraries
import numpy as np
from keras.preprocessing import image
from keras.applications import vgg16
import matplotlib.pyplot as plt
```

```python
#create model with vgg16
model = vgg16.VGG16()
```

```python
#load an image
#224*224 pixel is required by this model
#normalize with preprocessing
image1 = image.load_img('img1.jpg', target_size = (224, 224))
example = image.img_to_array(image1)
example = np.expand_dims(example, axis = 0)
example = vgg16.preprocess_input(example)
```

```python
#predict type of image
example_pr = model.predict(example)
types = vgg16.decode_predictions(example_pr)
```

```python
#show result
plt.imshow(image1)
plt.axis('off')
plt.show()
print('\n')

for imagenet_id, name, likelihood in types[0]:
    print("Prediction: {} - with proboblity of {:2f}".format(name, likelihood))
```
