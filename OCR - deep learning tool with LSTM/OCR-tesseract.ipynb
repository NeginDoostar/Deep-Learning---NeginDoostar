```python
# libraries
import cv2
import pytesseract
from matplotlib import pyplot as plt
```

```python
# load an image
image = cv2.imread(r'C:\Users\Negin\Pictures\Screenshots\text.png')

plt.figure(figsize = (7, 7))
plt.imshow(image)
plt.show()
# result of OCR
pytesseract.image_to_string(image)
```

# start preprocessing

```python
# change to gray scale, because of less noise and information and more intensity
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.figure(figsize = (7, 7))
plt.imshow(image)
plt.show()
# result of OCR
pytesseract.image_to_string(image)
```

```python
# use blure method to remove noise
cv2.medianBlur(image,5)
plt.figure(figsize = (7, 7))
plt.imshow(image)
plt.show()
# result of OCR
pytesseract.image_to_string(image)
```

```python
# edge detection with candy method to noise reduction, easier processing, boundary detection and improving accuracy
image = cv2.Canny(image, 100, 200)
plt.figure(figsize = (7, 7))
plt.imshow(image)
plt.show()
# result of OCR
pytesseract.image_to_string(image)
```
